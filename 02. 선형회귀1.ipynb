## **1.환경준비**

* 라이브러리 로딩

```
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

# 머신러닝 라이브러리
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
```


## **2. 데이터 준비**

### (1) data loading

```
path = 'https://raw.githubusercontent.com/DA4BAM/dataset/master/Carseats.csv'

use_cols = ['Sales', 'Price', 'Income', 'Advertising','CompPrice', 'ShelveLoc']

data = pd.read_csv(path, usecols= use_cols)
```

|	변수명	|	설명	|	구분	|
|----|----|----|
|	Sales 	|	 각 지역 판매량(단위 : 1000개)	|	Target	|
|	CompPrice 	|	지역별 경쟁사 판매가격(달러)	|	feature	|
|	Income 	|	가구당 평균 소득액(1000달러)	|	feature	|
|	Advertising 	|	 각 지역, 회사의 광고 예산(1000달러)	|	feature	|
|	Population 	|	 지역 인구수(단위 : 1000명)	|	feature	|
|	Price 	|	 자사 지역별 판매가격(달러)	|	feature	|
|	ShelveLoc 	|	 진열상태(범주 : Bad, Medium, Good)	|	feature	|
|	Age 	|	 지역 인구의 평균 연령	|	feature	|
|	Education 	|	 교육수준(범주 : 10~18)	|	feature	|
|	Urban 	|	 매장이 도심에 있는지 여부(범주 : Yes, No)	|	feature	|
|	US 	|	 매장이 미국에 있는지 여부(범주 : Yes, No)	|	feature	|


* 둘러보기

```
data.head()
```

### (2) 데이터분할1 : x, y 나누기

```
target = 'Sales'
x = data.drop(target, axis=1)   # axis=1 target이 칼럼
y = data.loc[:, target]
```

### (3) 가변수화

```
cat_cols = ['ShelveLoc'] # 범주형 데이터(위에 .info()에서 확인)
```

```
# 전체 가변수 사용
x_cat1 = pd.get_dummies(x, columns=cat_cols)
x_cat1.head()
```

```
# 첫번째 가변수 제거(첫번째 칼럼을 지워도 나머지로 설명 가능) -> drop_first=True
x_cat2 = pd.get_dummies(x, columns=cat_cols, drop_first=True)
x_cat2.head()
```

### (4) 데이터분할2 : train : validation 나누기

```
x_train, x_val, y_train, y_val = train_test_split(x_cat2, y, test_size = .3, random_state = 20)   # random_state 쓰면 매번 분할할 때 같게 나눠짐 -> 현업에서는 x
```

## 3.모델링 연습

```
from sklearn.linear_model import LinearRegression
```

```
model = LinearRegression()
```

```
model.fit(x_train, y_train)
```

```
pred = model.predict(x_val)
```

```
print(mean_absolute_error(y_val, pred))
```

## 4.모델링 : 단순회귀

* 하나의 feature로 tareget 예측
    * Price --> Sales

### (1)학습할 데이터 준비

```
features = ['Price']

x_train1 = x_train[features] # 위에서 만든 x_train에서 Price 칼럼만
x_val1 = x_val[features]
```

```
col = 'Price'
col_list = ['Price']

temp1 = x_train[col]        # Series
temp2 = x_train[col_list]   # DataFrame

print(type(temp1), type(temp2))
```

```
x_train1.head()
```

### (2) 모델선언

```
model1 = LinearRegression()
```

### (3) 모델링(학습)

```
model1.fit(x_train1, y_train)
```

* 회귀계수 확인하기

```
print('회귀계수 : ', model1.coef_)     ##.coef_: _는 값을 나타낼 때
print('절편 : ', model1.intercept_)
```

### (4) 검증 : 예측

```
pred1 = model1.predict(x_val1)
```

### (5) 검증 : 평가

* 평가용 함수들은 대부분 (실제값, 예측값)

```
# 평균오차
mean_absolute_error(y_val   , pred1   )
```

## 5.회귀 모델 평가하기

### (1) 오차의 비로 평가하기 : R2 Score

```
r2_score(y_val, pred1)       # #R^2 Score = SSR/SST = 1-SSE/SST
```

### (2)오차의 양과 율로 평가하기

```
# MSE
mean_squared_error( y_val   , pred1   )
```

```
# RMSE
root_mean_squared_error( y_val   , pred1)
```

```
# MAE
mean_absolute_error(y_val   , pred1)
```

```
# MAPE
mean_absolute_percentage_error(y_val   , pred1)
```

### (1) 단순회귀 모델1 : Income --> Sales

```
features2 = ['Income']

x_train2 = x_train[features2]
x_val2 = x_val[features2]
```

```
model2 = LinearRegression()

model2.fit(x_train2, y_train)
```

```
pred2 = model2.predict(x_val2)
```

```
# 평균오차
print("평균오차:", mean_absolute_error(y_val, pred2))
# R^2 Score(결정계수, 설명력)
print("r2_score:", r2_score(y_val, pred2))       # R^2 Score = SSR/SST = 1-SSE/SST
# MSE
print("MSE:", mean_squared_error(y_val, pred2))
# RMSE
print("RMSE:", root_mean_squared_error(y_val, pred2))
# MAE
print("MAE:", mean_absolute_error(y_val, pred2))
# MAPE
print("MAPE:", mean_absolute_percentage_error(y_val, pred2))
```
